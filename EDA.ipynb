```python
#Load libraries
import pandas as pd

#Import data
data = pd.read_csv("gtd.csv")
```

```python
#Some Exploratory Data Analysis

print("Year with the highest number of attacks: ", data['year'].value_counts().idxmax())
print("Month with the highest number of attacks:", data['month'].value_counts().idxmax(),"(May)")
print("Country with the highest number of attacks: ", data['country'].value_counts().idxmax())
print("Region with the highest number of attacks: ", data['region'].value_counts().idxmax())
print("State with the highest number of attacks: ", data['state'].value_counts().idxmax())
print("Success rate: ", round((data['success'].sum()/data['success'].count())*100, 2),"%")
print("Rate of suicide attack: ", round((data['suicide'].sum()/data['suicide'].count())*100, 2),"%")
print("Trending attack type: ", data['attacktype'].value_counts().idxmax())
print("Trending target type: ", data['targettype'].value_counts().idxmax())
print("Target Nationality: ", data['tar_nationality'].value_counts().idxmax())
print("Organization: ", data['organization'].value_counts().idxmax(), "-> Most attacks are unclaimed")
print("Civilian casuality: ", data['nkill'].sum())
print("Terrorist casuality: ", data['nkillter'].sum())
print("Wounded civilians: ", data['nwound'].sum())
print("Wounded terrorists: ", data['nwoundter'].sum())
```
